```python
import numpy as np
import numpy.polynomial.polynomial as npp
from matplotlib import pyplot as plt
```

```python
data_y = np.array([4.1,3.5,5.6,9.8,8.4,8.1,12.5,14.2,10.3,8.1,7.1,8.6])
#train_t = train_t.reshape([12, 1])
data_x = np.array(range(1, 13))
```

```python
data_y
```

```python
data_x
```

```python
plt.plot(data_x,data_y,'or')
```

```python
for order in range(2,6):
    fpoly = npp.polyfit(data_x, data_y, order)
    fpoly1d = np.poly1d(np.fliplr([fpoly])[0])
    print (fpoly1d)
    
    xPlot = np.arange(min(data_x)-1.0 , max(data_x)+1.0 , 0.1)
    yPlot = npp.polyval(xPlot, fpoly)
        
    plt.plot(xPlot, yPlot)
    plt.plot(data_x,data_y,'or')
    plt.show()
    plt.clf()
```
